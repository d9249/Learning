```python
import cv2

for idx in range(len(train)) :
    img = train.loc[idx, '0':].values.reshape(28, 28).astype(int)
    digit = train.loc[idx, 'digit']
    cv2.imwrite(f'./images_train/{digit}/{train["id"][idx]}.png', img)
```

```python
import cv2

for idx in range(len(test)) :
    img = test.loc[idx, '0':].values.reshape(28, 28).astype(int)
    cv2.imwrite(f'./images_test/{test["id"][idx]}.png', img)
```

```python
import tensorflow as tf
EfficientNetB7_model =  tf.keras.applications.EfficientNetB7(weights=None, include_top=True, input_shape=(224, 224, 1), classes=10)
```

```python
from tensorflow.keras.optimizers import Adam
EfficientNetB7_model.compile(loss='categorical_crossentropy', optimizer=Adam(lr=0.002,epsilon=None), metrics=['accuracy'])
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2,
                             rotation_range=10,
                             width_shift_range=0.1,
                             height_shift_range=0.1)

train_generator = datagen.flow_from_directory('./images_train', target_size=(224,224), color_mode='grayscale', class_mode='categorical', subset='training')
val_generator = datagen.flow_from_directory('./images_train', target_size=(224,224), color_mode='grayscale', class_mode='categorical', subset='validation')
```

```python
checkpoint = tf.keras.callbacks.ModelCheckpoint(f'/content/drive/MyDrive/DACON_CVLC/Checkpoint/CVLC_06_EfficientNetB7.h5', monitor='val_accuracy', save_best_only=True, verbose=1)
```

```python
EfficientNetB7_model.summary()
```

```python
EfficientNetB7_model.fit_generator(train_generator, epochs=500, validation_data=val_generator, callbacks=[checkpoint])
```

```python
import matplotlib.pyplot as plt

plt.plot(EfficientNetB7_model.history.history["accuracy"], label='EfficientNetB7_acc')
plt.plot(EfficientNetB7_model.history.history["val_accuracy"], label='EfficientNetB7_val')

plt.legend()
plt.show()
```

```python
datagen = ImageDataGenerator(rescale=1./255)
test_generator = datagen.flow_from_directory('./images_test', target_size=(224,224), color_mode='grayscale', class_mode='categorical', shuffle=False)
```

```python
EfficientNetB7_predict = EfficientNetB7_model.predict_generator(test_generator).argmax(axis=1)
```

```python
submission = pd.read_csv('/content/drive/MyDrive/DACON_CVLC/data/submission.csv')
submission.head()
```

```python
import numpy as np
mylist = []

for i in range(len(submission)):
    name =  test_generator.filenames
    id = name[i].split('/')[1].rstrip('.').split('.')[0]
    mylist.append(id)
print(mylist)
```

```python
for i in range(len(submission)):
    submission["id"][i] = mylist[i]
```

```python
submission["EfficientNetB7_predict"] = EfficientNetB7_predict
submission.head()
```

```python
from collections import Counter

for i in range(len(submission)) :
    predicts = submission.loc[i, ['EfficientNetB7_predict']]
    submission.at[i, "digit"] = Counter(predicts).most_common(n=1)[0][0]

submission.head()
```

```python
submission = submission[['id', 'digit']]
submission.head()
```
