# 6강. 그래프 그리기

데이터 특징, 분석의 결과, 현재 상태 등에 다양한 분석을 위해 시각화는 매우 중요한 요소 중에 하나다. 따라서 여러 가지 시각화 라이브러리가 있지만 기본적으로 널리 사용되는 Matplotlib을 알아본다. Matplotlib은 매우 많은 기능을 제공하고 있기 때문에 특별한 그래프나 디테일한 세팅이 필요하다면 구글링을 통해 해결하는 습관을 만들어 놓는 것이 좋다.

```python
import numpy as np
from matplotlib import pyplot as plt
```

## 6.1 Plot

```python
epoch = np.arange(1,100)
train_loss = 1.5/(epoch)
val_loss = .3+1/(epoch)
```

```python
plt.figure(figsize=(10,5))
plt.plot(train_loss, 'r-*')
plt.plot(val_loss, 'b-*')
plt.legend(['train','validation'])
plt.title("MNIST")
plt.xlabel("epoch")
plt.ylabel("loss")
plt.show()
```

```python
plt.figure(figsize=(5,10))
t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'ro', t, t**2, 'bs', t, 5*np.sin(t), 'g^')
plt.legend(['t','t**2','5sin(t)'])
plt.show()
```

## 6.2 Subplot

```python
names = ['First', 'Business', 'Economy']
prices = [20, 5, 1]

plt.figure(figsize=(15, 5))

plt.subplot(131)
plt.bar(names, prices)
plt.title(f"{names[0]}", fontsize=14)

plt.subplot(132)
plt.scatter(names, prices)
plt.xlabel('class', fontsize=12)

plt.subplot(133)
plt.plot(names, prices, linewidth=5.0)
plt.suptitle('Ticket Class', fontsize=14)
plt.axis('off')
plt.show()
```
