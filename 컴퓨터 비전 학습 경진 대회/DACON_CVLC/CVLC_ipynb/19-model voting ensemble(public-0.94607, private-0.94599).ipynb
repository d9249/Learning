```
submission = pd.read_csv('/content/drive/MyDrive/DACON_CVLC/data/submission.csv')
```

```
import numpy as np
mylist = []

for i in range(len(submission)):
    name =  test_generator.filenames
    id = name[i].split('/')[1].rstrip('.').split('.')[0]
    mylist.append(id)
```

```
for i in range(len(submission)):
    submission["id"][i] = mylist[i]
```

```
submission["ResNet50_predict_224"] = ResNet50_predict_224
submission["ResNet101_predict_224"] = ResNet101_predict_224
submission["ResNet152_predict_224"] = ResNet152_predict_224
submission["ResNet50V2_predict_224"] = ResNet50V2_predict_224
submission["ResNet101V2_predict_224"] = ResNet101V2_predict_224
submission["ResNet152V2_predict_224"] = ResNet152V2_predict_224
submission["InceptionV3_predict_224"] = InceptionV3_predict_224
submission["InceptionResNetV2_predict_224"] = InceptionResNetV2_predict_224
submission["DenseNet121_predict_224"] = DenseNet121_predict_224
submission["DenseNet169_predict_224"] = DenseNet169_predict_224
submission["DenseNet201_predict_224"] = DenseNet201_predict_224
submission["Xception_predict_224"] = Xception_predict_224
submission["EfficientNetB0_predict_224"] = EfficientNetB0_predict_224
submission["EfficientNetB1_predict_224"] = EfficientNetB1_predict_224
submission["EfficientNetB2_predict_224"] = EfficientNetB2_predict_224
submission["EfficientNetB3_predict_224"] = EfficientNetB3_predict_224
submission["EfficientNetB4_predict_224"] = EfficientNetB4_predict_224
submission["EfficientNetB5_predict_224"] = EfficientNetB5_predict_224
submission["EfficientNetB6_predict_224"] = EfficientNetB6_predict_224
```

```
from collections import Counter

for i in range(len(submission)) :
    predicts = submission.loc[i, ['ResNet50_predict_224',
                                  'ResNet101_predict_224',
                                  'ResNet152_predict_224',
                                  'ResNet50V2_predict_224',
                                  'ResNet101V2_predict_224',
                                  'ResNet152V2_predict_224',
                                  'InceptionV3_predict_224',
                                  'InceptionResNetV2_predict_224',
                                  'DenseNet121_predict_224',
                                  'DenseNet169_predict_224',
                                  'DenseNet201_predict_224',
                                  'Xception_predict_224',
                                  'EfficientNetB0_predict_224',
                                  'EfficientNetB1_predict_224',
                                  'EfficientNetB1_predict_224',
                                  'EfficientNetB2_predict_224',
                                  'EfficientNetB3_predict_224',
                                  'EfficientNetB4_predict_224',
                                  'EfficientNetB5_predict_224',
                                  'EfficientNetB6_predict_224'
                                  ]]
    submission.at[i, "digit"] = Counter(predicts).most_common(n=1)[0][0]
```

```
submission = submission[['id', 'digit']]
```
