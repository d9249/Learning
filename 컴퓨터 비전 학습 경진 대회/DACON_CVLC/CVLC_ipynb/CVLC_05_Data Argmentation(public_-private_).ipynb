```
# one hot encoder
def one_hot_encoder(x):
    y = np.zeros((len(x), len(np.unique(x))))  # 모든 값이 0인 백터
    for i, num in enumerate(x):
        y[i][num] = 1  # Label에 해당하는 인덱스에 1을 입력
    return y
```
